```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

df = pd.read_csv('data_vs_som.csv', sep=',',decimal='.')
df.head()
```

```python
for size in df['size'].unique() :
    disps=df['disp'].unique() 
    colors = cm.rainbow(np.linspace(0, 1, len(disps)))
    df_size=df[df['size']==size]
    df_size=df_size.sort_values('som_size')

    fig, ax = plt.subplots(figsize = (12, 8)) 

    for disp, c in zip(disps,colors):
        df_disp=df_size.loc[df_size['disp'] == disp]
        ax.plot(df_disp['som_size'], df_disp['unused_som'], color = c , marker = 'o', label = disp)
        #for i, txt in enumerate(h1['img_name']):
        #    ax.annotate(txt, (h1[['focus']].iloc[i], h1['measure'].iloc[i]))

    ax.set_title(f'data size of {size}',fontsize=24) #Ttol grfic
    ax.set_xlabel('som_size',fontsize=22) #Etiqueta eix x
    ax.set_ylabel('%_unused_som',fontsize=22) #Etiqueta eix y
    #ax.set_xlim((550, 1050)) 
    #ax.set_ylim((0, 1))
    ax.legend(loc = 'upper left',fontsize=18) 
    ax.grid(True)
    ax.tick_params(axis='both', which='major', labelsize=16)
```
